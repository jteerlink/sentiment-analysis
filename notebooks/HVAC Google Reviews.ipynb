```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine
import urllib
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

params = urllib.parse.quote_plus('DRIVER={SQL Server Native Client 10.0};Server=gmlsql2hou; DATABASE=invplannersapp;Trusted_Connection=yes;')
engine = create_engine("mssql+pyodbc:///?odbc_connect=%s" % params)
```

```python
df = pd.read_sql("select * from invplannersapp.dbo.HVAC_Reviews",con=engine,parse_dates=['ParsedDate'])
```

```python
df['AverageRating']=df['AverageRating'].astype('float',copy=False)
df['Rating']=df['Rating'].astype('float',copy=False)
df['TotalReviews']=df['TotalReviews'].astype('int',copy=False)
```

```python
analyser = SentimentIntensityAnalyzer()
```

```python
temp_list = []
for i in df['ReviewText']:
    try:
        score = analyser.polarity_scores(i)
        temp_list.append(score)
    except:
        temp_list.append({'compound':0,'neg':0,'neu':0,'pos':0})
        

sentiment = pd.DataFrame(temp_list)
```

```python
final = pd.merge(df,sentiment,left_index=True, right_index=True)
```

```python
final[final['TotalReviews']>=5].groupby(final.Name).mean().sort_values(by='compound',ascending=False)
```
